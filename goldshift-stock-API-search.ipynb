```python
import sys
import urllib.parse
import requests
from alpha_vantage.timeseries import TimeSeries 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np

```

```python
API_key = '5DSE3C5SKVGXPHNX'
ts = TimeSeries(key=API_key, output_format='pandas')
companies = ['BVN', 'GOLD', 'NEM', 'KGC', 'HMY', 'CDE', 'GFI']
```

```python


def stock_comps_API(companies, ts):
    merged_data = pd.DataFrame()  
    
    for ticker in companies:
        data, meta_data = ts.get_daily(symbol=ticker, outputsize='full')
        close_prices = data['4. close']
        close_prices.name = ticker
        merged_data = pd.concat([merged_data, close_prices], axis=1)
        
    merged_data.index = pd.to_datetime(merged_data.index)
    data_range = pd.bdate_range(start='2004-01-01', end='2024-07-01')
    valid_dates = data_range.intersection(merged_data.index)
    merged_data = merged_data.loc[valid_dates]
    return merged_data

# Call the function
goldshift_df = stock_comps_API(companies, ts)
```
